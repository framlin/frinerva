# comput the tenent service charge payments sum for 2020
- Read bank statement 2020
- filter service charge entires for 2020
- sum them up

## ==>  Caution  <==
### Rechnungabgrenzung:
Zahlungen für **Januar 2021** wurden ausgefiltert und müssen in Nebenkostenabrechnung 2021 berücksichtigt werden


```python
from accounting.cost_center_mapping import cost_center_mapping as ccm
```

```python
df_bookings = pd.read_csv('../data/imports/2020_MANUAL.csv', delimiter=';')
```

```python
def convert_amount(data):
    return float(data['Betrag'].replace(',', '.'))

def map_category(data):
    return ccm[data['Kategorie'].split(' - ')[-1]]

def filter_tenant_payments(data):
    return ( convert_amount(data) > 0) \
            and (map_category(data) in {'SERVICE_CHARGES': 1, }.keys())\
            and (data['Verwendungszweck'].find('Januar') == -1)

    
tenant_payments = pd.DataFrame().from_records([{'description' : data['Verwendungszweck'], 'amount': convert_amount(data)} \
                                               for _, data in df_bookings.iterrows() if filter_tenant_payments(data)])
tenant_payments_sum = tenant_payments['amount'].sum()
```

### Nebenkostenvorauszahlung Mieter

```python
print(tenant_payments)
md('#### Total amount: %i €'%(tenant_payments_sum))
```

```python
def filter_landloard_payments(data):
    return ( convert_amount(data) < 0) \
            and (map_category(data) in {'SERVICE_CHARGES': 1, }.keys())\
            and (data['Verwendungszweck'].find('Quartal') == -1)\
            and (data['Verwendungszweck'].find('Grundsteuer') == -1)

landlord_payments = pd.DataFrame().from_records([{'description' : data['Verwendungszweck'], 'amount': convert_amount(data)} \
                                               for _, data in df_bookings.iterrows() if filter_landloard_payments(data)])
landlord_payments_sum = landlord_payments['amount'].sum()
```
